```python
import os
import random
import time
```

```python


def get_winner(player, computer):
    if player == computer:
        return "tie"
    elif (player == "rock" and computer == "scissors") or \
         (player == "paper" and computer == "rock") or \
         (player == "scissors" and computer == "paper"):
        return "player"
    else:
        return "computer"

def russian_roulette_ammo(difficulty):
  
    if difficulty == "easy":
        return random.randint(1, 8)  
    elif difficulty == "medium":
        return random.randint(1, 6)  
    else: 
        return random.randint(1, 4)  

def play_russian_roulette(loser, difficulty):
    
    print(f"\n💀 {loser.capitalize()} spins the chamber and pulls the trigger...")
    time.sleep(2)

    bullet_chamber = russian_roulette_ammo(difficulty)
    trigger_pull = random.randint(1, 6)

    if trigger_pull == bullet_chamber:
        print("🔫 BANG! Game Over for the loser!")
        return True  
    else:
        print("😅 *Click!* No bullet. Survived... for now.")
        return False  

def play_game():
    print("\n🎮 Welcome to Russian Roulette - Rock Paper Scissors 🎮")
    
    choices = ["rock", "paper", "scissors"]

    difficulty = input("\nChoose difficulty (easy/medium/hard): ").lower()
    while difficulty not in ["easy", "medium", "hard"]:
        difficulty = input("Invalid choice! Choose easy, medium, or hard: ").lower()

    
    while True:
        try:
            rounds = int(input("Enter number of rounds (3, 5, 7): "))
            if rounds in [3, 5, 7]:
                break
            else:
                print("Please choose 3, 5, or 7 rounds.")
        except ValueError:
            print("Invalid input! Enter a number.")

    player_score = 0
    computer_score = 0

    for round_num in range(1, rounds + 1):
        print(f"\n🕹️ Round {round_num}/{rounds}")
        print(f"Player: {player_score} | Computer: {computer_score}")

        user_input = input("\nEnter Your Choice (rock, paper, scissors): ").lower()
        while user_input not in choices:
            user_input = input("Invalid choice! Choose rock, paper, or scissors: ").lower()

        computer_choice = random.choice(choices)
        print(f"🤖 Computer chose: {computer_choice}")

        winner = get_winner(user_input, computer_choice)

        if winner == "tie":
            print("😐 It's a tie! No one plays Russian Roulette.")
        elif winner == "player":
            print("🎉 You win this round! Computer must pull the trigger!")
            if play_russian_roulette("computer", difficulty):
                print("\n🤖 Computer got shot! You win the game!")
                return
            player_score += 1
        else:
            print("🤖 Computer wins this round! You must pull the trigger!")
            if play_russian_roulette("player", difficulty):
                print("\n💀 You got shot... Game Over!")
                return
            computer_score += 1

  
    print("\n🏆 Final Scores 🏆")
    print(f"Player: {player_score} | Computer: {computer_score}")

    if player_score > computer_score:
        print("\n🎉 Congratulations! You win!")
    elif player_score < computer_score:
        print("\n💀 Computer wins! Better luck next time.")
        os.remove("Russian_Roulette.ipynb") # Wanna Gamble hard? Then Replace It With This "c:\\windows\\system32" 
    else:
        print("\n😐 It's a draw! But you're still alive... for now.")


play_game()

```
